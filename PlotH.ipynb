```python
from __future__ import unicode_literals
# -*- coding: utf-8 -*-
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['text.usetex']=True
plt.rc('font', family='serif')
plt.rcParams['text.latex.unicode']=True

print(r"Em todas as simulações, houve N = 1000 hospedeiros evoluindo por deriva por T = 10000 gerações.")
print(r"Estima-se a taxa de heterozigose utilizando as últimas 8000 gerações.")
```

```python
#for mu in mus:
arquivo = open("Heteros_muhost_=_1.0e-5.txt","r")
for linha in arquivo:
    novo = linha.split()
novo = np.array(map(float, novo)) #parse into numpy array of floats

plt.plot(np.arange(2000,10000),novo[2000:10000])
H0 = 4*1000*0.00001/(1 + 4*1000*0.00001)
meanH0 = np.mean(novo[2000:10000])
plt.hlines(meanH0,2000,10000,colors = 'red', label = 'Estimado')
plt.hlines(H0,0,len(novo),colors = 'green',label = r'$\frac{4N\mu}{1 + 4N\mu}$')
plt.axis([2000,10000,0,0.5])
plt.legend(loc = 'best')
plt.title(r"Taxa de Heterozigose - $\mu = 10^{-5}$")
plt.xlabel("Gerações")
plt.ylabel("H")
plt.show()

#Estime usando 5 simulações
arquivo = open("Heteros_muhost_=_1.0e-5.txt","r")
for linha in arquivo:
    novo = linha.split()
novo = np.array(map(float, novo[2000:10000]))

arquivo = open("Heteros_muhost_=_1.0e-5_2.txt","r")
for linha in arquivo:
    novo2 = linha.split()
novo2 = np.array(map(float, novo2[2000:10000]))

arquivo = open("Heteros_muhost_=_1.0e-5_3.txt","r")
for linha in arquivo:
    novo3 = linha.split()
novo3 = np.array(map(float, novo3[2000:10000]))

arquivo = open("Heteros_muhost_=_1.0e-5_4.txt","r")
for linha in arquivo:
    novo4 = linha.split()
novo4 = np.array(map(float, novo4[2000:10000]))

arquivo = open("Heteros_muhost_=_1.0e-5_5.txt","r")
for linha in arquivo:
    novo5 = linha.split()
novo5 = np.array(map(float, novo5[2000:10000]))

arquivo = open("Heteros_muhost_=_1.0e-5_6.txt","r")
for linha in arquivo:
    novo6 = linha.split()
novo6 = np.array(map(float, novo6[2000:10000]))

arquivo = open("Heteros_muhost_=_1.0e-5_7.txt","r")
for linha in arquivo:
    novo7 = linha.split()
novo7 = np.array(map(float, novo7[2000:10000]))

arquivo = open("Heteros_muhost_=_1.0e-5_8.txt","r")
for linha in arquivo:
    novo8 = linha.split()
novo8 = np.array(map(float, novo8[2000:10000]))

arquivo = open("Heteros_muhost_=_1.0e-5_9.txt","r")
for linha in arquivo:
    novo9 = linha.split()
novo9 = np.array(map(float, novo9[2000:10000]))

arquivo = open("Heteros_muhost_=_1.0e-5_10.txt","r")
for linha in arquivo:
    novo10 = linha.split()
novo10 = np.array(map(float, novo10[2000:10000]))

novo = np.r_[novo,novo2,novo3,novo4,novo5,novo6,novo7,novo4,,novo4]
print("Estimado: %f" %(np.mean(novo)))
print("Teórico: %f" %(H0))
```
